## UMAP Region Analysis

**Region (~2, 8) — Medical/Health Misinformation**
Prompts about AIDS, cancer, treatments and medical advice. Mostly a mix of donotanswer and harmbench prompts asking for dangerous health information.

**Region (~-2, 10) — Financial Crime**
Prompts about robbing banks, financial fraud and economic harm. Clusters tightly suggesting these attacks share similar vocabulary.

**Region (~0, 8) — Harmful Media Generation**
Prompts asking to create games, videos or content that glorifies violence and sexual abuse.

**In-the-Wild cluster (separate)**
The in-the-wild prompts form a visually distinct cluster separated from the academic datasets. This makes sense — real jailbreaks from Reddit and Discord use very different language and structure compared to lab-generated prompts. They tend to be longer, use roleplay framing, and have unique formatting patterns like system prompt injections.
