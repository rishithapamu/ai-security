```python
import sys
import os
sys.path.insert(0, os.path.abspath('../..'))

import pandas as pd
import numpy as np

# Load cluster results
df = pd.read_parquet('../data/clusters/clusters.parquet')
print(f"Total records: {len(df)}")
print(f"Clusters: {df['cluster'].nunique()}")
print(f"Noise points: {(df['cluster'] == -1).sum()}")
```

```python
# Show 10 random prompts from each cluster for labeling
import random

cluster_ids = sorted([c for c in df['cluster'].unique() if c != -1])

for cluster_id in cluster_ids:
    cluster_df = df[df['cluster'] == cluster_id]
    sample = cluster_df['prompt'].sample(min(10, len(cluster_df)), random_state=42)
    top_category = cluster_df['attack_category'].value_counts().index[0] \
        if cluster_df['attack_category'].notna().any() else "unknown"
    
    print(f"\n{'='*60}")
    print(f"CLUSTER {cluster_id} | {len(cluster_df)} points | top category: {top_category}")
    print('='*60)
    for prompt in sample:
        truncated = f"{prompt[:300]}..." if len(prompt) > 300 else prompt
        print(f" - {truncated}")
        
```
